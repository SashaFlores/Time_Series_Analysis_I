# Time Series Basics

```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Read the Amazon stock prices CSV file as a time series DataFrame 
# Be sure to set the `Date` column as the index
# YOUR CODE HERE!
amazon = pd.read_csv(Path('../Resources/amazon.csv'), parse_dates=True, index_col='Date')
amazon.head(2)
```

### Slice Time Series Data

```python
# Select all rows from September 2018
# YOUR CODE HERE!
amazon_2018 = amazon['sep. 2018']
amazon_2018
```

```python
# Select all rows from September through October for 2018
# YOUR CODE HERE!
sep_oct = amazon['2018-09' : '2018-10']
sep_oct
```

### Plot Time Series Data

```python
# Plot the closing prices using a line plot
# YOUR CODE HERE!
amazon.Close.plot(title='Amazon Closing Prices')
```

### Resample Time Series Data

```python
# Resample the closing prices to weekly and take the mean
# YOUR CODE HERE!
weekly = amazon.resample('W').mean()
weekly.head(2)
```

```python
# Plot the weekly average closing prices as a line chart
# YOUR CODE HERE!
weekly.plot(title='Weekly Average Clsoing Prices')
```
